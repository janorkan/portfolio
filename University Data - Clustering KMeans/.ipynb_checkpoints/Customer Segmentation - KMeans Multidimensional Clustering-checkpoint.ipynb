# Customer Segmentation - KMeans Multidimensional Clustering
In this ML project we are using the KMeans Clustering to segment customers based on a Kaggle dataset.

```python
import numpy as np
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt
import kaggle
import kagglehub
from kagglehub import KaggleDatasetAdapter
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from summarytools import dfSummary
from matplotlib.colors import ListedColormap
from mpl_toolkits.mplot3d import Axes3D
import warnings
%matplotlib inline
```

## 1. Data and Visualization

```python
# Download data from Kaggle
df = kagglehub.dataset_load(KaggleDatasetAdapter.PANDAS,"vjchoudhary7/customer-segmentation-tutorial-in-python","Mall_Customers.csv",)

dfSummary(df)
```

```python
# 2D Visualization: Income vs Spending Score
sns.set_style("whitegrid")
sns.scatterplot(df, x="Annual Income (k$)", y="Spending Score (1-100)")
plt.title("Income vs Spending Score")
```

```python
# Clean and convert categorical data into discrete numbers
female = {"Female": 1, "Male":0}

df_clean = df.copy(deep=True)

df_clean.drop(["CustomerID"], axis=1, inplace=True)

df_clean["Female"] = df_clean["Gender"].map(female)
df_clean.drop(["Gender"], axis=1, inplace=True)

df_clean.head()
```

```python
# 3D Visualization: Income vs Spending Score vs Age 
fig = plt.figure(figsize=(5,5))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
cmap = ListedColormap(sns.color_palette("nipy_spectral"))

threeD = ax.scatter(xs=df_clean["Annual Income (k$)"], ys=df_clean["Spending Score (1-100)"], zs=df_clean["Age"], c=df_clean["Female"] , cmap=cmap)
ax.set_title("Income vs Spending Score vs Age")
ax.set_xlabel("Annual Income")
ax.set_ylabel("Spending Score")
ax.set_zlabel("Age")
```

```python
# Scaling data

scaler = StandardScaler()
df_scale = pd.DataFrame(scaler.fit_transform(df_clean), columns=["Age","Annual Income (k$)", "Spending Score (1-100)","Female"])
df_scale.head()
```

## 2. 2D Clustering: Annual Income & Spending Score
Using the inertia from the ellbow method and the silhouette_score to identify the optimal cluster number.

In the first approach we are using only the two columns Annual Income & Spending Score.

```python
df_scale_2D = df_scale[["Annual Income (k$)","Spending Score (1-100)"]]
df_scale_2D.head()
```

```python
# Find best cluster number
warnings.filterwarnings('ignore')

max_cluster = 10
inertia_2D = {}
silhouette_2D = {}
labels_2D = {}

for i in range(1,max_cluster):
    kmeans = KMeans(n_clusters = i, random_state = 42, init="k-means++")
    kmeans.fit(df_scale_2D)
    
    inertia_2D[i] = kmeans.inertia_
    labels_2D[i] = kmeans.labels_
    if i >= 2:
        silhouette_2D[i] = silhouette_score(df_scale_2D, kmeans.fit_predict(df_scale_2D))

print(inertia_2D)
print(silhouette_2D)                                 
```

```python
# Visualize Ellbow-Curve

sns.lineplot(x = inertia_2D.keys(), y= inertia_2D.values(), marker="o", color="red")
plt.xlabel("K values")
plt.ylabel("Inertia")
plt.title("Ellbow Curve")
```

```python
# Visualize Silhouette Score

sns.set_style("whitegrid")
sns.lineplot(x = silhouette_2D.keys(), y= silhouette_2D.values(), marker="o", color="green")
plt.xlabel("K values")
plt.ylabel("Score")
plt.title("Silhouette Score")
```

```python
# Cluster Data for K=5

df["2D Clusters K=5"] = labels_2D[5]
df
```

```python
# Visualization
sns.scatterplot(df, x="Annual Income (k$)", y="Spending Score (1-100)", hue="2D Clusters K=5", palette="nipy_spectral")
```

## 3. 3D Clustering: Annual Income, Spending Score and Age
Using the inertia from the ellbow method and the silhouette_score to identify the optimal cluster number.

Adding "Age" as a 3rd feature dimension for clustering.

```python
df_scale_3D = df_scale[["Annual Income (k$)","Spending Score (1-100)", "Age"]]
df_scale_3D.head()
```

```python
# Find best cluster number
warnings.filterwarnings('ignore')

inertia_3D = {}
silhouette_3D = {}
labels_3D = {}

for i in range(1,max_cluster):
    kmeans = KMeans(n_clusters = i, random_state = 42, init="k-means++")
    kmeans.fit(df_scale_3D)
    
    inertia_3D[i] = kmeans.inertia_
    labels_3D[i] = kmeans.labels_
    if i >= 2:
        silhouette_3D[i] = silhouette_score(df_scale_3D, kmeans.fit_predict(df_scale_3D))

print(inertia_3D)
print(silhouette_3D)                                 
```

```python
# Visualize Ellbow-Curve

sns.lineplot(x = inertia_3D.keys(), y= inertia_3D.values(), marker="o", color="orange")
plt.xlabel("K values")
plt.ylabel("Inertia")
plt.title("Ellbow Curve")
```

```python
# Visualize Silhouette Score

sns.set_style("whitegrid")
sns.lineplot(x = silhouette_3D.keys(), y= silhouette_3D.values(), marker="o", color="green")
plt.xlabel("K values")
plt.ylabel("Score")
plt.title("Silhouette Score")
```

```python
# Cluster Data for K=5

df["3D Clusters K=6"] = labels_3D[6]
df
```

```python
# 3D Clustering: Income vs Spending Score vs Age 
fig = plt.figure(figsize=(5,10))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
cmap = ListedColormap(sns.color_palette("nipy_spectral"))

threeD = ax.scatter(xs=df["Annual Income (k$)"], ys=df["Spending Score (1-100)"], zs=df["Age"], c=df["3D Clusters K=6"] , cmap=cmap)
ax.set_title("Income vs Spending Score vs Age")
ax.set_xlabel("Annual Income")
ax.set_ylabel("Spending Score")
ax.set_zlabel("Age")
```

## 4. N-dimensional Clustering

Next we are adding the reminding dimension "Gender" to the clustering.

```python
# Find best cluster number
warnings.filterwarnings('ignore')

max_cluster_n = 15
inertia_nD = {}
silhouette_nD = {}
labels_nD = {}

for i in range(1,max_cluster_n):
    kmeans = KMeans(n_clusters = i, random_state = 42, init="k-means++")
    kmeans.fit(df_scale)
    
    inertia_nD[i] = kmeans.inertia_
    labels_nD[i] = kmeans.labels_
    if i >= 2:
        silhouette_nD[i] = silhouette_score(df_scale, kmeans.fit_predict(df_scale))

print(inertia_nD)
print(silhouette_nD)                                 
```

```python
# Visualize Ellbow-Curve

sns.lineplot(x = inertia_nD.keys(), y= inertia_nD.values(), marker="o", color="purple")
plt.xlabel("K values")
plt.ylabel("Inertia")
plt.title("Ellbow Curve")
```

```python
# Visualize Silhouette Score

sns.set_style("whitegrid")
sns.lineplot(x = silhouette_nD.keys(), y= silhouette_nD.values(), marker="o", color="green")
plt.xlabel("K values")
plt.ylabel("Score")
plt.title("Silhouette Score")
```

```python
# Cluster Data for K=11

df["nD Clusters K=11"] = labels_nD[11]
df
```

## 5. N-dimensional Clustering with PCA

As the results for the n-dimensional clustering were not that useful, we are doing a dimension reduction with PCA.

```python
pca = PCA(n_components=2)

df_PCA = pd.DataFrame(pca.fit_transform(df_scale),columns=['PCA1','PCA2'])
```

```python
# Find best cluster number
warnings.filterwarnings('ignore')

inertia_PCA = {}
silhouette_PCA = {}
labels_PCA = {}

for i in range(1,max_cluster):
    kmeans = KMeans(n_clusters = i, random_state = 42, init="k-means++")
    kmeans.fit(df_PCA)
    
    inertia_PCA[i] = kmeans.inertia_
    labels_PCA[i] = kmeans.labels_
    if i >= 2:
        silhouette_PCA[i] = silhouette_score(df_PCA, kmeans.fit_predict(df_PCA))

print(inertia_PCA)
print(silhouette_PCA)                                 
```

```python
# Visualize Ellbow-Curve

sns.lineplot(x = inertia_PCA.keys(), y= inertia_PCA.values(), marker="o", color="pink")
plt.xlabel("K values")
plt.ylabel("Inertia")
plt.title("Ellbow Curve")
```

```python
# Visualize Silhouette Score

sns.set_style("whitegrid")
sns.lineplot(x = silhouette_PCA.keys(), y= silhouette_PCA.values(), marker="o", color="green")
plt.xlabel("K values")
plt.ylabel("Score")
plt.title("Silhouette Score")
```

```python
# Cluster Data for K=4

df["2D PCA Clusters K=4"] = labels_PCA[4]
df_PCA["2D PCA Clusters K=4"] = labels_PCA[4]
df
```

```python
sns.scatterplot(df_PCA, x="PCA1", y="PCA2", hue="2D PCA Clusters K=4", palette="nipy_spectral")
plt.title("Clustering in 2-PCA Dimensions")
```

```python
# 3D Clustering: Income vs Spending Score vs Age 
fig = plt.figure(figsize=(5,5))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
cmap = ListedColormap(sns.color_palette("nipy_spectral"))

threeD = ax.scatter(xs=df["Annual Income (k$)"], ys=df["Spending Score (1-100)"], zs=df["Age"], c=df["2D PCA Clusters K=4"] , cmap=cmap)
ax.set_title("2d PCA Income vs Spending Score vs Age")
ax.set_xlabel("Annual Income")
ax.set_ylabel("Spending Score")
ax.set_zlabel("Age")
```

# Summary
In this Kmeans customer clustering approach we walked from a 2-dimensional clustering, over a 3d and 4d clustering into a PCA-reduced 2d clustering.

While the 2d clustering delivered easy to understand and clearly seperated data with a high silhouette score (> 0.5) and 5 clusters, the 3d and 4d clustering delivered clusterings with lower silhouette score and more flexibel cluster numbers. Adding the age nearly doubled the number of suggested clusters in this dataset.

The 2-dimensional PCA approach at the end helped to improve the n-dimensional KMeans clustering, but still has low silhouette scores and a weak ellbow curve.
