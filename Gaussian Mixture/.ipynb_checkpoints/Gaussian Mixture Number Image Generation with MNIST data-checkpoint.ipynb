# Gaussian Mixture Number Image Generation with MNIST data
In this project we are using the MNIST dataset, which contains over 700 digitalized images of handwritten numbers. The goal is to generate new number images with the GaussianMixture model. 

```python
import sklearn.mixture
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import os

os.environ["OMP_NUM_THREADS"] = '1'
```

## 0. Load and Visualize MNIST Data

```python
# load and show dataset
mnist = datasets.load_digits()
mnist
```

```python
mnist.images[0].shape
```

```python
# Visualize single images
_, axes = plt.subplots(ncols=8, figsize=(15, 3))
for  ax, image, label in zip(axes, mnist.images, mnist.target):
    ax.imshow(image, cmap=plt.cm.gray_r,  interpolation="nearest")
    ax.axis('off')
    ax.set_title("Number: %i" % label)
```

## 1. Gaussian Mixture

```python
X = mnist.data

# Normalize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce dimensionality with PCA
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Gaussian_Mixture
gm = GaussianMixture(n_components=10, covariance_type='full', random_state=42)
gm.fit(X_pca)

# Sample data                     
samples, labels = gm.sample(n_samples=64)

# Retransformation
samples = pca.inverse_transform(samples)                     
samples = scaler.inverse_transform(samples)
```

```python
# Get shape of images
samples[0].shape
```

```python
# Reshape images
samples_shape = samples.reshape(-1, 8, 8)
samples_shape[0].shape
```

```python
# Plot images
fig, axes = plt.subplots(nrows=8, ncols=8, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    ax.imshow(samples_shape[i], interpolation="nearest", cmap='gray')
    ax.axis('off')
plt.show()
```

# 2. Improve Number Generation
In the previous step we trained the GaussianMixture with alle images for all numbers. This can be improved by training the GaussianMixture only with images from one number. Therefore we split the data by the numbers and train the GaussianMixture with the specific numbers.

```python
X = pd.DataFrame(mnist.data)
y = pd.DataFrame(mnist.target).rename(columns={0:"target"})

X2 = pd.concat([X,y], axis=1, join="inner")
X2["target"].value_counts()
```

```python
# Select only 3s and drop target column
X3 = X2[X2["target"] == 3].drop(["target"], axis=1)
X3
```

```python
# Normalize data
X3_scaled = scaler.fit_transform(X3)

# Reduce dimensionality with PCA
X3_pca = pca.fit_transform(X3_scaled)

# Gaussian_Mixture
gm3 = GaussianMixture(n_components=10, covariance_type='full', random_state=42)
gm3.fit(X3_pca)

# Sample data                     
samples3, labels3 = gm3.sample(n_samples=64)

# Retransformation
samples3 = pca.inverse_transform(samples3)                     
samples3 = scaler.inverse_transform(samples3)

# Reshape images
samples3_shape = samples3.reshape(-1, 8, 8)
samples3_shape[0].shape

# Plot images
fig, axes = plt.subplots(nrows=8, ncols=8, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    ax.imshow(samples3_shape[i], interpolation="nearest", cmap='gray')
    ax.axis('off')
plt.show()
```
