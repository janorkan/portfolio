# Financial Data Analysis

```python
import os
import requests
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from summarytools import dfSummary
import datetime
%matplotlib inline
```

## Load Data
We are using the FMP api to retrieve 

```python
API_KEY = os.getenv("FMP_API_KEY", "API error")
```

```python
# Stock List
                   
stocks_url = f"https://financialmodelingprep.com/api/v3/stock/list?&apikey={API_KEY}"
stocks_response = requests.get(stocks_url)

if stocks_response.status_code == 200:
    stock_list = stocks_response.json()
    df_stocks = pd.DataFrame(stock_list)

    df_stocks.to_csv("stock_list.csv", index=False)
    
    print(f"Stock list retrieved! {len(df_stocks)} stocks saved to 'stock_list.csv'.")
else:
    print(f"Error fetching stock list! Status code: {stocks_response.status_code}")
    df_stocks = pd.DataFrame()  # Empty DataFrame as fallback
```

```python
# Ticker target list

companies = {
    "Rheinmetall AG [EUR]": "RNMBF",
    "BAE Systems [USD]": "BAESF",
    "Hensoldt [EUR]": "HAGHY",
    "Thales Group [EUR]": "THLEF",
    "Leonardo S.p.A. [EUR]": "FINMF",
    "Lockheed Martin [USD]": "LMT"
}
```

```python
# Historical Stock prices

end_date   = datetime.datetime.today().strftime("%Y-%m-%d")
start_date = (datetime.datetime.today() - datetime.timedelta(weeks=260)).strftime("%Y-%m-%d")
data_dict = {}

for company, ticker in companies.items():
    hist_url = f"https://financialmodelingprep.com/api/v3/historical-price-full/{ticker}?from={start_date}&to={end_date}&apikey={API_KEY}"
    hist_response = requests.get(hist_url)
    
    if hist_response.status_code == 200:
        hist_data = hist_response.json()
        if "historical" in hist_data and hist_data["historical"]:
            df_hist = pd.DataFrame(hist_data["historical"])
            df_hist["date"] = pd.to_datetime(df_hist["date"])
            df_hist.sort_values("date", inplace=True)
            df_hist.set_index("date", inplace=True)
            data_dict[company] = df_hist["close"]
        else:
            print(f"Keine historischen Daten fÃ¼r {company} ({ticker}) gefunden.")
    else:
        print(f"Fehler beim Abrufen der historischen Daten fÃ¼r {company} ({ticker}). Status code: {hist_response.status_code}")
```

```python
# Convert dict to parallelized dataframe

df_stock_prices = pd.DataFrame.from_dict(data_dict)
df_stock_prices
```

## Plotting and Reindexing

```python
# Plotting stock prices
sns.set_style("darkgrid")
plt.figure(figsize=(20,8))
sns.lineplot(df_stock_prices, palette="nipy_spectral")
```

```python
# Reindexing on 2023-01-01 (1-year after start of Russian Invasion in Ukraine)

index_date = "2023-02-22"

df_stock_prices_index = df_stock_prices[index_date:index_date]
df_stock_prices_indexed = pd.DataFrame(columns=df_stock_prices.columns)

for col in df_stock_prices_index:
    df_stock_prices_indexed[col]=df_stock_prices[col]/df_stock_prices_index[col].iloc[0]*100

df_stock_prices_indexed.loc[index_date:]
```

```python
# Plotting indexed stock prices

plt.figure(figsize=(20,8))
sns.lineplot(df_stock_prices_indexed[index_date:], palette="nipy_spectral")
```

## Trailing Returns

```python
# Trailing Returns

time_offsets = {
    "1Y Return %": pd.DateOffset(years=1),
    "3M Return %": pd.DateOffset(months=3),
    "1M Return %": pd.DateOffset(months=1),
    "1W Return %": pd.DateOffset(weeks=1),
}
latest_date = df_stock_prices.index[-1]
latest_prices = df_stock_prices.loc[latest_date]

# Get latest price
def get_closest_price(df,date):
    available_dates = df.index[df.index <= date]
    return df.loc[available_dates[-1]] if len(available_dates) > 0 else None  

# Iterate Offsets
returns = {}
for label, offset in time_offsets.items():
    past_date = latest_date - offset
    past_price = get_closest_price(df_stock_prices,past_date)
    
    if past_price is not None:
        returns[label] = round((latest_prices / past_price - 1)*100,2)
    else:
        returns[label] = None

# Ergebnisse als DataFrame anzeigen
df_returns = pd.DataFrame(returns).T
df_returns
```

```python
#Barplots for each Period
returns_df = df_returns.T

fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(10, 10))
axes = axes.flatten()

for i, offset in enumerate(returns_df.columns):
    sns.barplot(x=returns_df.index, y=returns_df[offset], ax=axes[i], hue = returns_df.index, palette="nipy_spectral") 
    axes[i].set_title(offset)
    axes[i].set_ylabel("Return (%)")
    axes[i].set_xlabel("Stocks")
    axes[i].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

### Moving Average: 30d & 200d

```python
df_stock_prices_roll30 = df_stock_prices.rolling(window=30, min_periods = 20).mean()
df_stock_prices_roll200 = df_stock_prices.rolling(window=200, min_periods=150).mean()

fig, axes = plt.subplots(nrows=len(df_stock_prices.columns), ncols=1, figsize=(10, 4 * len(df_stock_prices.columns)), sharex=True)

for i, stock in enumerate(df_stock_prices.columns):
    ax = axes[i] 

    ax.plot(df_stock_prices.index, df_stock_prices[stock], label=f"{stock} Price", color='black', linewidth=1)

    ax.plot(df_stock_prices.index, df_stock_prices_roll30[stock], label="30-Day MA", color='blue', linestyle="dashed")
    
    ax.plot(df_stock_prices.index, df_stock_prices_roll200[stock], label="200-Day MA", color='red', linestyle="dashed")
    
    ax.set_title(f"{stock} - Price & Moving Averages")
    ax.set_ylabel("Price")
    ax.legend()

plt.xlabel("Date")
plt.tight_layout()
plt.show()
```
