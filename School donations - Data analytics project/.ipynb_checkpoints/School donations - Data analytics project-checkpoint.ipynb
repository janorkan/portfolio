## 1. Data Import & Basic Analysis
The underlying data are splitted in 6 csv files, extracted from a realtional database with primary and secondarys identifiers. We import these files as panda dataframes.
```python
# Import all files in project data path
all_files =  Path('./data').glob('*.csv')
dfs = []

try:
    for f in all_files:
        df = pd.read_csv(f, index_col=None, header=0)
        df.name = f.stem
        dfs.append(df)
except IOError:
    print('Failed to load data!')
```

### 3.4. Identify Donation increase by Student-Led Projects
In the previous analysis we identified, that stduent-led projects have significant higher donations per project and per donation itself. In the next analysis we want to find out, if it is impossible to identify reasons for the increased spending. Therefore we will have a closer look at the donor,school and project information.

```python
# Grouping by categories to get datasheet for correlation calculation
df_ptype_deep = df_3.groupby(['Project Type','Donor Is Teacher', 'School Metro Type', 'Project Grade Level Category', 'Project Resource Category', 'School State', 'Teacher Prefix']).agg({'Donation Amount': 'sum', 'Project ID': 'nunique', 'Donation ID': 'nunique'}).rename(columns={'Donation ID':'# Donations', 'Project ID':'# Projects'})
df_ptype_deep.reset_index(inplace=True)

df_ptype_deep
```

```python
# LabelEncoding of categorical values

df_ptype_deep_le =df_ptype_deep.drop(columns=(['Donation Amount','# Projects','# Donations']),axis=1).apply(LabelEncoder().fit_transform)

df_ptype_deep_le = pd.concat([df_ptype_deep_le,df_ptype_deep[['Donation Amount','# Projects','# Donations']]], axis=1, join='inner')

df_ptype_deep_le
```

```python
# Correlation (Spearman)

df_ptype_corr = df_ptype_deep_le.corr(method='spearman').round(2)
df_ptype_corr
```

```python
fig = px.imshow(df_ptype_corr, color_continuous_scale='Oranges', title='Spearman Correlations')
fig.layout.height = 500
fig.update_layout(title_text='Spearman Correlations', title_x=0.5)
fig.show()
```

```python
# Definitions of Xicor (Author: https://gist.github.com/itamarfaran/0cffce51cd0e3d78ddf509b72a9e702d)
import warnings
warnings.filterwarnings('ignore', '^overflow encountered in scalar .*', category=RuntimeWarning) 

def xicor(x, y, ties='auto'):
    x = np.asarray(x,dtype=np.float64).flatten()
    y = np.asarray(y,dtype=np.float64).flatten()
    n = len(y)

    if len(x) != n:
        raise IndexError(
            f"x, y length mismatch: {len(x)}, {len(y)}"
        )

    if ties == 'auto':
        ties = len(np.unique(y)) < n
    elif not isinstance(ties, bool):
        raise ValueError(
            f"expected ties either \"auto\" or boolean, "
            f"got {ties} ({type(ties)}) instead"
        )
    
    y = y[np.argsort(x)]
    r = rankdata(y, method='ordinal')
    nominator = np.sum(np.abs(np.diff(r)))

    if ties:
        l = rankdata(y, method='max')
        denominator = 2 * np.sum(l * (n - l))
        nominator *= n
    else:
        denominator = np.power(n, 2) - 1
        nominator *= 3

    statistic = 1 - nominator / denominator  # upper bound is (n - 2) / (n + 1)
    p_value = norm.sf(statistic, scale=2 / 5 / np.sqrt(n))

    return statistic, p_value
    
xic1 = xicor(df_ptype_deep_le['Donor Is Teacher'],df_ptype_deep_le['Donation Amount'])
print(xic1[0])
```

```python
# Calcluate Xicor Correlation Matrix

def xicor_corr_matrix(df):
    col_names = df.columns
    num_cols = len(col_names)
    result = np.zeros((num_cols, num_cols))
    
    for i in range(num_cols):
        for j in range(num_cols):
            result[i, j] = xicor(df.iloc[:, i], df.iloc[:, j])[0]
    
    return pd.DataFrame(result, index=col_names, columns=col_names)

df_xicor_corr = xicor_corr_matrix(df_ptype_deep_le)
df_xicor_corr
```

**Discussion of Correlation Results**
- We LabelEncoded all categorical columns to calculate a numerical correlation. This encoding is random, so even smaller amounts of correlation can give a hint for real correlations.
- We used the Spearmen correlation instead of the usual linear Person correlation first, because of the randomness of the label-encoding. There are 3 interesting categories with a higher correlation than 0.1, we can dive in: "Project Resource Categorie", "Donation is Teacher" and the "Project Type", we already identified.
- As an alternatively and more flexible correaltion, we used the state-of-the-art xicor correlation function and built a custom Xicor correlation matrix. Values near 1 show higher correlation, than values close to zero and much higher than 1.
- Besides the 3 mentioned categories above, we should also look into the "School States".

```python
# Project type vs Donor is Teacher: donation amount

df_ptype_teacher = df_3.groupby(['Project Type','Donor Is Teacher']).agg({'Donation Amount': 'sum', 'Project ID': 'nunique', 'Donation ID': 'nunique'}).rename(columns={'Donation ID':'# Donations', 'Project ID':'# Projects'})
df_ptype_teacher.reset_index(inplace=True)
df_ptype_teacher = df_ptype_teacher.pivot(index='Project Type', columns='Donor Is Teacher', values=['Donation Amount','# Projects','# Donations'])

df_ptype_teacher.reset_index(inplace=True)
df_ptype_teacher['% Teacher Donation Amount'] = df_ptype_teacher.loc[:,('Donation Amount','Yes')]/df_ptype_teacher.loc[:,('Donation Amount','No')]*100
df_ptype_teacher
```

```python
# Project Type vs Project Resource Categorie

df_ptype_resource = df_3.groupby(['Project Type','Project Resource Category']).agg({'Donation Amount': 'sum'})
df_ptype_resource.reset_index(inplace=True)

df_ptype_resource = df_ptype_resource.pivot(columns='Project Type', index='Project Resource Category', values='Donation Amount')
df_ptype_resource['% Student-Led'] = df_ptype_resource.loc[:,'Student-Led']/(df_ptype_resource.loc[:,'Professional Development'] + df_ptype_resource.loc[:,'Teacher-Led'])*100

df_temp = df_3.groupby(['Project Resource Category']).agg({'Donation Amount': 'sum','Donation ID': 'nunique'})
df_temp['Amount per Donation'] = round(df_temp.loc[:,'Donation Amount']/df_temp.loc[:,'Donation ID'],2)

df_ptype_resource = pd.concat([df_ptype_resource,df_temp['Amount per Donation']],axis=1)

df_ptype_resource_sort = df_ptype_resource[['Amount per Donation','% Student-Led']].sort_values('Amount per Donation', ascending=False)
df_ptype_resource_sort
```

```python
df_ptype_resource_sort.iplot(kind='scatter', x='Amount per Donation', y='% Student-Led',  yTitle='% Student-Led', xTitle='Donation Amount', title ='Student-Led Share per Donation Amount', colorscale='oranges')
```

```python
df_ptype_resource_sort.corr()
```

**Result Discussion**
- In this examplatory analysis step, we looked deeper into teacher donations and donation resource types.
- While teacher donate a lot to professional projects, they donate lesser to student-led and teacher-led projects.
- A bigger correlation exists between resource types and the donation amount for student-led projects.
- Student-led projects are more often higher donated resource types like trips, lab equipment and so.
