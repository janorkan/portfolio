# Polynomial Regression with K-Fold Crossvalidation
In this project, we are looking into polynomial regression. Therefore, we create random data and identify the best fitting polynom with the K-Folding crossvalidation method.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

## Create Data and Visualize Polynoms


```python
# Create and visualize data

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

polgrade1 = np.poly1d(np.polyfit(x,y,1))
polgrade2 = np.poly1d(np.polyfit(x,y,2))
polgrade3 = np.poly1d(np.polyfit(x,y,3))
polgrade4 = np.poly1d(np.polyfit(x,y,4))

myline = np.linspace(1,25,100)
plt.scatter(x,y)
plt.plot(myline, polgrade1(myline), label = "Grade 1")
plt.plot(myline, polgrade2(myline), label = "Grade 2")
plt.plot(myline, polgrade3(myline), label = "Grade 3")
plt.plot(myline, polgrade4(myline), label = "Grade 4")
plt.legend()
plt.show()
```

## K-Fold Validation using sklearn

```python
# Defining Kfold Cross validation method for polynomial features, using the LinearRegression model for regression method behind the PolynomialFeature transformation

def kfold_cross_validation_polynomials(X, y, degrees, k):
    X = np.array(X)
    y = np.array(y)
    kfold = KFold(n_splits=k, shuffle=True, random_state=42)
    ms_errors = {degree: [] for degree in degrees}
    ms_sum_errors = {degree: [] for degree in degrees}

    for degree in degrees:
        poly_feature = PolynomialFeatures(degree=degree)
        for train_i, test_i in kfold.split(X):
            X_train, X_test = X[train_i], X[test_i]
            y_train, y_test = y[train_i], y[test_i]
            
            X_train_pf = poly_feature.fit_transform(X_train)
            X_test_pf = poly_feature.transform(X_test)
            
            lin_model = LinearRegression()
            lin_model.fit(X_train_pf, y_train)
            y_predict = lin_model.predict(X_test_pf)
            
            ms_error = mean_squared_error(y_test, y_predict)
            ms_errors[degree].append(ms_error)

        ms_sum_error = sum(ms_errors[degree])
        ms_sum_errors[degree].append(ms_sum_error)

    return ms_sum_errors
```

```python
# Identify best polynomial grade for given dataset
x = np.array(x)
x = x.reshape(-1,1)

kfold_errors = kfold_cross_validation_polynomials(x,y,[1,2,3,4],10)
print(kfold_errors)
```

# Summary
In this small project, we did a polynomial linear regression with the mean_squared_error to find the ideal polynomial grade with the mean squared error method using kfold cross validation with the help of the sklearn library.

for teh given data, a 3-grade polynomial linear regression has the lowest mean_square_error, closely followed by a 4-grade polynom. These two candidates should be considered for further machine learning approaches in this scenario.
