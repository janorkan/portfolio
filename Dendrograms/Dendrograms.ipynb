# Dendrograms
In this project we are visualizing all available distance methods to calculate the linakge_matrix for dendrograms used for hierarchical clustering.

Therefore we can use the scipy libraries dendrogram and linkage.

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
# Data

x = [4, 6, 9, 4, 3, 11, 12 , 6, 10, 12]

y = [22, 18, 25, 16, 16, 24, 24, 22, 21, 21]
```

```python
# Create 2 dimensional data list

data = list(zip(x,y))
print(data)
```

```python
# Calculate Linkage and Visualize dendrograms

distance_methods = ["single", "complete", "average", "weighted", "centroid", "median", "ward"]

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

for i, method in enumerate(distance_methods):
    linkage_matrix = linkage(data, method=method)
    dendrogram(linkage_matrix, ax=axes[i])
    axes[i].set_title(f"Dendrogram - {method}")
```
