# Faker - Synthetic User Data
In this exercise I'm generating a DataFrame full of authentic german user data using the faker library.

```python
import pandas as pd
from faker import Faker
```

```python
df = pd.DataFrame(columns=["USER_ID", "FIRSTNAME", "LASTNAME", "STREET", "HOUSENUMBER", "ZIPCODE", "CITY", "LAT", "LON", "PREFERENCE", "CURRENCY", "TIMEZONE"])

fake = Faker("de_DE")

word_list = ["loves", "hates", "buyes", "does not buy", "steals", "borrows", "Salat", "Steak", "Carots"]

for row in range(0,10):
    fake_results = []
    fake_results.append(fake.unique.random_int(min=1, max=100))
    fake_results.append(fake.unique.first_name())
    fake_results.append(fake.last_name())
    fake_results.append(fake.street_name())
    fake_results.append(fake.random_int(min=1, max=127))
    fake_results.append(fake.postcode())
    fake_results.append(fake.city_name())  
    fake_results.append(fake.latitude())
    fake_results.append(fake.longitude())
    fake_results.append(fake.sentence(ext_word_list=word_list))
    fake_results.append(fake.currency_symbol())
    fake_results.append(fake.timezone())
    df.loc[len(df)] = fake_results

df.head(10)
```
