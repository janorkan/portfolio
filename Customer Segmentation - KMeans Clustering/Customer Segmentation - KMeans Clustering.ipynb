# Customer Segmentation - KMeans Clustering
In this ML project we are using the KMeans Clustering to segment customers by two features:
1. Age
2. Avg Order_Sum

The data are generated randomly for test purposes only. The following approach is not restricted to two features, but chosen for easier visualization and understanding.

```python
import numpy as np
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import warnings
```

```python
# Load and Visualize data
df = pd.read_csv("customer_segmentation.csv")

sns.set_style("whitegrid")
sns.scatterplot(df, x="Age", y="Average_Order_Sum")
plt.title("Customer Data")
```

## 2. Identify Optimal Cluster Number
Using the inertia from the ellbow method and the silhouette_score to identify the optimal cluster number.

```python
# Scale data

scaler = StandardScaler()
df_scale = scaler.fit_transform(df)
```

```python
# Find best cluster number
warnings.filterwarnings('ignore')

max_cluster = 10
inertia = {}
silhouette = {}
labels = {}

for i in range(1,max_cluster):
    kmeans = KMeans(n_clusters = i, random_state = 42, init="k-means++")
    kmeans.fit(df_scale)
    
    inertia[i] = kmeans.inertia_
    labels[i] = kmeans.labels_
    if i >= 2:
        silhouette[i] = silhouette_score(df_scale, kmeans.fit_predict(df_scale))

print(inertia)
print(silhouette)                                 
```

```python
# Visualize Ellbow-Curve

sns.lineplot(x = inertia.keys(), y= inertia.values(), marker="o", color="red")
plt.xlabel("K values")
plt.ylabel("Inertia")
plt.title("Ellbow Curve")
```

```python
# Visualize Silhouette Score

sns.set_style("whitegrid")
sns.lineplot(x = silhouette.keys(), y= silhouette.values(), marker="o", color="green")
plt.xlabel("K values")
plt.ylabel("Score")
plt.title("Silhouette Score")
```

## 3. Visualize Data with Clusters
The ellbow curve and the silhouette score are suggesting either 4 or 5 clusters. Lets's visualize them.

```python
# Cluster Data

df["K=4 Clusters"] = labels[4]
df["K=5 Clusters"] = labels[5]
df
```

```python
# Visualization

fig,ax = plt.subplots(1,2, figsize=(10,6))
fig.tight_layout()
g1 = sns.scatterplot(df, x="Age", y="Average_Order_Sum", hue="K=4 Clusters", palette="nipy_spectral", ax=ax[0])
g1.legend([],[], frameon=False)
ax[0].set_title("K=4 Clusters")
g2 = sns.scatterplot(df, x="Age", y="Average_Order_Sum", hue="K=5 Clusters", palette="nipy_spectral", ax=ax[1])
g2.legend([],[], frameon=False)
ax[1].set_title("K=5 Clusters")
```

# Summary
In this test ML project for KMeans clustering we identified the best number of clusters for the generated dataset. Due to the different scaled features we scaled them for a meaningful euclidic distance calculation. If you don't do that, the Average_Order_Sum feature will dominate the clustering.

To find the optimal cluster number, we calculated the inertia and the silhouette score for up to 10 different clusters. The results of this mathematical approach suggests 4/5 clusters for the dataset. In the end this would be a business decision, if they want to split the upper-left cluster into two, or not.

This approach can be extended from 2 features up-to n features without any bigger adjustments, except the plotting. For multiple features a dimension reduction technic like PCA should be added to identify the principal components for clustering.
